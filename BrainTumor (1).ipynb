```python
import os
import glob
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Dense, Dropout, Conv2D, MaxPooling2D, Flatten
from keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from keras.optimizers import RMSprop
from sklearn.model_selection import train_test_split

# Function to Load Image Paths
def import_data(path):
    return [os.path.abspath(filename) for filename in glob.glob(os.path.join(path, '*'))]

# Paths
path_no = "/content/extracted_files/no"
path_yes = "/content/extracted_files/yes"
path_pred = "/content/extracted_files/pred"

# Load Data
train_no = import_data(path_no)
train_yes = import_data(path_yes)
test_images = import_data(path_pred)

# Create DataFrames
df_train_no = pd.DataFrame({'image': train_no, 'label': 'Healthy'})
df_train_yes = pd.DataFrame({'image': train_yes, 'label': 'Affected'})
df_test = pd.DataFrame({'image': test_images})

# Combine Training Data
train_data = pd.concat([df_train_no, df_train_yes])

# Convert Labels to Strings (Fix for class_mode="sparse")
train_data['label'] = train_data['label'].astype(str)

# Train-Validation Split
X_train, X_val = train_test_split(train_data, test_size=0.1, shuffle=True, random_state=42)

# CNN Model Function
def build_model():
    model = Sequential([
        Conv2D(64, (5, 5), activation='relu', padding='same', strides=(2, 2), input_shape=(224, 224, 1)),
        MaxPooling2D(2),
        Conv2D(128, (5, 5), activation='relu', padding='same', strides=(2, 2)),
        Conv2D(128, (5, 5), activation='relu', padding='same', strides=(2, 2)),
        Conv2D(256, (5, 5), activation='relu', padding='same', strides=(2, 2)),
        MaxPooling2D(2),
        Flatten(),
        Dense(64, activation='relu'),
        Dropout(0.4),
        Dense(32, activation='relu'),
        Dropout(0.4),
        Dense(2, activation='softmax')
    ])
    return model

# Training Function
def train_model(train_data, val_data):
    model = build_model()
    model.compile(optimizer=RMSprop(learning_rate=1e-4), loss='sparse_categorical_crossentropy', metrics=['acc'])

    callbacks = [
        EarlyStopping(monitor='val_loss', mode='min', patience=3, restore_best_weights=True, verbose=1),
        ModelCheckpoint("Brain_model_best.keras", save_best_only=True),
        ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-5, mode='min', verbose=1)
    ]

    history = model.fit(train_data, validation_data=val_data, epochs=50, batch_size=10, callbacks=callbacks)

    # Save model for future use
    model.save("Brain_tumor_model.h5")
    print("Model saved successfully.")

    return history

# Cross-Validation Training
def cross_validation_training(train_data, val_data, k_fold=3):
    datagen = ImageDataGenerator(rescale=1./255)
    cv_histories = []

    for i in range(k_fold):
        train_set = datagen.flow_from_dataframe(train_data, x_col='image', y_col='label', target_size=(224, 224),
                                                color_mode='grayscale', class_mode='sparse', batch_size=10, shuffle=True)
        val_set = datagen.flow_from_dataframe(val_data, x_col='image', y_col='label', target_size=(224, 224),
                                              color_mode='grayscale', class_mode='sparse', batch_size=10, shuffle=True)
        print(f"Training on Fold: {i + 1}")
        cv_histories.append(train_model(train_set, val_set))

    return cv_histories

# Train the Model
cv_results = cross_validation_training(X_train, X_val)

# Function to Display Accuracy and Loss Plots
def plot_results(results):
    for fold in results:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
        fig.suptitle("Model's Metrics Visualization")

        ax1.plot(fold.history['acc'])
        ax1.plot(fold.history['val_acc'])
        ax1.set_title('Accuracy')
        ax1.set_xlabel('Epochs')
        ax1.set_ylabel('Accuracy')
        ax1.legend(['Training', 'Validation'])

        ax2.plot(fold.history['loss'])
        ax2.plot(fold.history['val_loss'])
        ax2.set_title('Loss')
        ax2.set_xlabel('Epochs')
        ax2.set_ylabel('Loss')
        ax2.legend(['Training', 'Validation'])

        plt.show()

plot_results(cv_results)

# Load Best Model
model = tf.keras.models.load_model('Brain_model_best.keras')

# Prediction on Test Set
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_dataframe(df_test, x_col='image', target_size=(224, 224),
                                            color_mode='grayscale', class_mode=None, batch_size=10, shuffle=False)

# Get Predictions
predictions = model.predict(test_set).argmax(axis=-1)
pred_labels = ['Healthy' if p == 1 else 'Affected' for p in predictions]

print(pred_labels)

# Display Predictions
def visualize_predictions(images, predictions):
    for img, pred in zip(images, predictions):
        plt.imshow(img.reshape(224, 224), cmap='gray')
        plt.title(f"Prediction: {pred}")
        plt.axis('off')
        plt.show()

# Fix Test Set Image Extraction
images = next(iter(test_set))
visualize_predictions(images, pred_labels[:10])
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
import os

# Model path (ensure the model exists)
model_path = r'/content/Brain_model_best.keras'  # Update this path

# Check if the model file exists
if not os.path.exists(model_path):
    print(f"Error: Model file '{model_path}' not found. Train the model first.")
    exit()

# Load the trained model
model = tf.keras.models.load_model(r'/content/Brain_tumor_model.h5')

# Function to Predict Tumor in MRI Image
def predict_mri(image_path):
    if not os.path.exists(image_path):
        print(f"Error: The file was not found at {image_path}. Check the file path and try again.")
        return

    # Load the image and preprocess
    img = image.load_img(image_path, target_size=(224, 224), color_mode='grayscale')
    img_array = image.img_to_array(img)
    img_array = np.expand_dims(img_array, axis=0)  # Add batch dimension
    img_array /= 255.0  # Normalize pixel values

    # Make prediction
    prediction = model.predict(img_array).argmax(axis=-1)[0]
    predicted_label = "Healthy" if prediction == 1 else "Affected"

    # Display the image and prediction
    plt.imshow(img_array[0].reshape(224, 224), cmap='gray')
    plt.title(f"Prediction: {predicted_label}")
    plt.axis('off')
    plt.show()

    print(f"Result: The MRI scan is classified as **{predicted_label}**.")

# Provide your MRI scan path here (FIXED PATH)
mri_image_path = r'/content/y3.jpg'  # Update with your image path

# Predict if the image has a brain tumor
predict_mri(mri_image_path)
```
